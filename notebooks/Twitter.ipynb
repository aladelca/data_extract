```python
from datetime import datetime
usuarios = []
tweet_id = []
contenido = []
imagen = []
lenguaje = []
fecha = []
users = {u['id']: u for u in response.includes['users']}
for tweet in response.data:
    if users[tweet.author_id]:
        user = users[tweet.author_id]
        id = tweet.id
        texto = tweet.text
        user_name = user.username
        profile_image = user.profile_image_url
        tweet_date = tweet.created_at.strftime("%Y-%m-%d %H:%M:%S")
        idioma = tweet.lang
        lenguaje.append(idioma)
        fecha.append(tweet_date)
        usuarios.append(user_name)
        tweet_id.append(id)
        contenido.append(texto)
        imagen.append(profile_image)


    
```

```python
import pandas as pd
data = pd.DataFrame(list(zip(tweet_id,usuarios,contenido,lenguaje,fecha,imagen)), columns = ['id','username','contenido','idioma','fecha','imagen_perfil'])
data_en = data[data['idioma'] == 'en']
data_en['contenido'] = data_en['contenido'].astype(str).str.lower()
data_en
```

```python
import nltk
nltk.download('vader_lexicon')
nltk.download('stopwords')
```

```python
from nltk.tokenize import RegexpTokenizer

regexp = RegexpTokenizer('\w+')

data_en['text_token'] = data_en['contenido'].apply(regexp.tokenize)
data_en.head(3)
```

```python
from nltk.corpus import stopwords
stopwords = nltk.corpus.stopwords.words("english")

# Extend the list with your own custom stopwords
my_stopwords = ['https','www']
stopwords.extend(my_stopwords)
```

```python
data_en['text_token'] = data_en['text_token'].apply(lambda x: [item for item in x if item not in stopwords])
data_en.head(3)
```

```python
data_en['text_complete'] = data_en['text_token'].apply(lambda x: ' '.join([item for item in x]))
data_en.head()
```

```python
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk.stem import WordNetLemmatizer

wordnet_lem = WordNetLemmatizer()

data_en['text_complete'] = data_en['text_complete'].apply(wordnet_lem.lemmatize)
data_en.head()
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()
```

```python
data_final = pd.concat([data_en,data_en['text_complete'].apply(lambda x: analyzer.polarity_scores(x)).apply(pd.Series)], axis = 1)
data_final.head()
```
